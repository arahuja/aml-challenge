```python
import pandas as pd

trainData = pd.read_csv('trainingData-release.csv')


testData = pd.read_csv('scoringData-release.csv')

print set(trainData.columns).difference(set(testData.columns))
```

```python
import numpy as np
from sklearn.decomposition import PCA, KernelPCA
from sklearn.preprocessing import scale

from sklearn.lda import LDA

markers = trainData[[   
'resp.simple',
'CD34', 
                    'CD7',
                                    'CD20',
                                    'HLA.DR', 
                                    'CD33', 
                                    'CD10',
                                    'CD13',
                                    'CD19',
                                    'ALBUMIN',
                                    'FIBRINOGEN',
                                    'BM.PROM',
                                    'PB.MONO',
                                    'PB.PROM',
                                    'BM.MONOCYTES']]

markers.dropna(inplace=True)


y = np.array(markers['resp.simple'].map(lambda x: 1 if x == 'CR' else 0))
markers.drop(['resp.simple'], axis=1, inplace=True)
X = scale(markers)
X
pca = KernelPCA(n_components = 2, kernel="rbf", fit_inverse_transform=True, gamma=3)
X_r = pca.fit_transform(X)


lda = LDA(n_components=2)
X_r2 = lda.fit(X, y).transform(X)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

target_names = ['resistant', 'remission']
plt.figure()
for c, i, target_name in zip("rgb", [0, 1], target_names):
    plt.scatter(X_r[y == i, 0], X_r[y == i, 1], c=c, label=target_name)
plt.legend()
plt.title('PCA of dataset')
```

```python
from sklearn.manifold import MDS

tnn = MDS(n_components=2)
X_tnn = tnn.fit_transform(X)

target_names = ['resistant', 'remission']
plt.figure()
for c, i, target_name in zip("rgb", [0, 1], target_names):
    plt.scatter(X_tnn[y == i, 0], X_tnn[y == i, 1], c=c, label=target_name)
plt.legend()
plt.title('tSNE of dataset')
```
